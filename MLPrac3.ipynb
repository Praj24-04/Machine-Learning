```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/StudentsPerformance.csv')
df.head()
```



## Box Plot

```python
plt.boxplot(df['math score'],
            patch_artist=True,
            boxprops=dict(facecolor='lightgreen'))

plt.title('Colored Box Plot')
plt.show()
```

```python
plt.boxplot([df['math score'],
             df['reading score'],
             df['writing score']],
            labels=['Math', 'Reading', 'Writing'])

plt.title('Multiple Box Plots')
plt.show()
```

##Scatter Plot

```python
colors = df['gender'].map({'female': 'pink', 'male': 'blue'})

plt.scatter(df['math score'], df['reading score'],
            c=colors)

plt.title('Scatter Plot by Gender')
plt.show()
```

## Bar plot

```python
avg_math_gender = df.groupby('gender')['math score'].mean()

plt.bar(avg_math_gender.index,
        avg_math_gender.values,
        color=['orange', 'green'])

plt.title('Bar Plot with Colors')
plt.show()
```

```python
plt.barh(avg_math_gender.index,
         avg_math_gender.values,
         color='skyblue')

plt.title('Horizontal Bar Plot')
plt.show()
```

##Pie Chart

```python
gender_counts = df['gender'].value_counts()

plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Gender Distribution of Students')
plt.show()
```

##Line Chart

```python
avg_scores = df[['math score', 'reading score', 'writing score']].mean()

plt.plot(avg_scores.index, avg_scores.values)
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.title('Average Scores by Subject')
plt.show()
```

```python
plt.plot(df['math score'][:20],
         label='Math',
         color='red',
         linestyle='--')

plt.plot(df['reading score'][:20],
         label='Reading',
         color='blue',
         linestyle='-.')

plt.plot(df['writing score'][:20],
         label='Writing',
         color='green',
         linestyle=':')

plt.legend()
plt.title('Multi-Line Styled Chart')
plt.show()
```
