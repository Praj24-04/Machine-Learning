Univarite Analysis

```python
def univariate_analysis(df):
    df = df.select_dtypes(include=np.number)
    result = pd.DataFrame({
        "Mean": df.mean(),
        "Median": df.median(),
        "Mode": df.mode().iloc[0],
        "Variance": df.var(),
        "Std Dev": df.std(),
        "Skewness": df.skew(),
        "Kurtosis": df.kurt()
    })
    return result
```

```python
uni_result = univariate_analysis(df)
uni_result
```
